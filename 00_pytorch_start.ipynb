### Tensor Datatypes

**Note**

Three common errors seen in pytorch and deep learning:
1. Tensors are not right datatype.
2. Tensors are not right shape.
3. Tensors are not on the right device.

```python
# Importent parameters can be used while creating a tensor
float_tensor = torch.tensor([3.0, 6.0, 8.0],
                            dtype=torch.float16,
                            device = None,        # "cpu"(by default), "cuda"(for gpu multithreading)
                            requires_grad=False)  # Wheather or not to track gradients with this tensor operations
float_tensor, float_tensor.dtype
```

```python
# converting the float16 type tensor to float32
float_32_tensor = float_tensor.type(torch.float32)
float_32_tensor, float_32_tensor.dtype
```

### Tensor Reshaping, stacking, squeezing and unsqueezing

```python
x = torch.arange(1,19)
x, x.dtype, x.ndim
```

```python
x = x.type(torch.float32)
x.dtype
```

```python
x_reshape = x.reshape(1, 7)
x_reshape, x_reshape.shape
```

```python
# Add an extra dimention
x_re = x.reshape(1, 18)
x_re, x_re.shape, x_re.ndim
```

```python
x_re = x.reshape(18, 1)
x_re, x_re.shape, x_re.ndim
```

```python
# 2 dimentional reshape
x_2d_re = x_re.reshape(9, 2)
x_2d_re, x_2d_re.shape, x_2d_re.ndim
```

```python
# Change the view
# similar to reshape (but reshape does not create a new copy of tensor it uses the memory of the input tensor itself)
# this is kind of a call by reference in c
T = torch.range(1, 9)
T.shape, T
```

```python
z = T.view(9, 1)
z.shape, z
```

```python
T.shape, T
```

```python
# stack tensors on top of each other
x_stacked_hstack = torch.stack([T,T,T,T], dim=0)
x_stacked_vstack = torch.stack([T,T,T,T], dim=1)
x_stacked_hstack, x_stacked_vstack, x_stacked_vstack.shape
```

```python
# torch.squeeze() - removes all single dimention from a target tensor
x_st_v_re = x_stacked_vstack.reshape(1, 9, 4)
x_st_v_re, x_st_v_re.ndim
```

```python
x_sqz = x_st_v_re.squeeze()
x_sqz, x_sqz.ndim
```

```python
# torch.unsqueeze() - adds a single dimention to a target tensor at a specific dimension.
x_unsqz = x_sqz.unsqueeze(dim=0)
x_unsqz, x_unsqz.shape, x_unsqz.ndim
```

```python
x_unsqz = x_sqz.unsqueeze(dim=1)
x_unsqz, x_unsqz.shape, x_unsqz.ndim
```

```python
x_unsqz = x_sqz.unsqueeze(dim=2)
x_unsqz, x_unsqz.shape, x_unsqz.ndim
```

### PyTorch Reproducbility (Trying to take random out of random)

```python
import torch
```

```python
# to reduce the randomness in neural networks and PyTorch comes the concept of a random seed
# Essentially what the random seed does is flavour the ransomness

rand_tensor_A = torch.rand(3, 4)
rand_tensor_B = torch.rand(3, 4)

rand_tensor_A, rand_tensor_B
```

```python
rand_tensor_A == rand_tensor_B
```

```python
# set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

rand_tensor_A = torch.rand(3, 4)
rand_tensor_B = torch.rand(3, 4)

rand_tensor_A, rand_tensor_B
```

```python
rand_tensor_A == rand_tensor_B
```

```python
# use the seed for every random number generating block of code
torch.manual_seed(RANDOM_SEED)
rand_tensor_A = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
rand_tensor_B = torch.rand(3, 4)

print(rand_tensor_A)
print(rand_tensor_B)

rand_tensor_A == rand_tensor_B
```

```python
# Create two random tensors of shape (2, 3)
# and send them both to the GPU (you'll need access to a GPU for this).
#Set torch.manual_seed(1234) when creating the tensors
# (this doesn't have to be the GPU random seed)

import torch
RANDOM_SEED = 1234
```

```python
torch.manual_seed(RANDOM_SEED)
T1 = torch.rand(2, 3)
T2 = torch.rand(2, 3)
T1, T1
```

```python
if torch.cuda.is_available():
  T1_GPU = T1.to('cuda')
  T2_GPU = T2.to('cuda')
  print('gpu')
else:
  print('cpu')
T1_GPU.device, T2_GPU.device
```

```python
# Perform a matrix multiplication on the tensors you created

output = torch.matmul(T1_GPU, T2_GPU.T)
output, output.size
```

```python
# Find the maximum and minimum values of the output
output_max = torch.max(output)
output_min = torch.min(output)
output_max, output_min
```

```python
# Find the maximum and minimum index values of the output
arg_max = torch.argmax(output)
arg_min = torch.argmin(output)
arg_max, arg_min
```
